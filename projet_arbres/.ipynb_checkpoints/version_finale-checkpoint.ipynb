```python
from numpy import *
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("les-arbres.csv",sep = ";",header = 0)
df.head(5)
```

# Creation de la dataframe  hauteur des arbres est comprise entre 1 et 40 m

```python
df_new =  df[(df['HAUTEUR']>=1) & (df['HAUTEUR']<=40)]
df_new.shape
```

# Hauteur maximum par arrondissement

```python
#Hauteur maximum par arrondissement 
plt.figure(figsize=(20,10))
max_arbre=df_new.groupby(['ARRONDISSEMENT']).HAUTEUR.max().reset_index(name='taille maximum')
plt.bar(max_arbre['ARRONDISSEMENT'],max_arbre['taille maximum'],color='salmon')
plt.xticks(rotation=45, ha='right',color='black')
print()
```

# Hauteur minimum par arrondissement 

```python
#hauteur minimum par arrondissement 
plt.figure(figsize=(20,10))
min_arbre = df_new.groupby(['ARRONDISSEMENT']).HAUTEUR.min().reset_index(name='taille minimum')
plt.bar(min_arbre['ARRONDISSEMENT'],min_arbre['taille minimum'],color='orange')
plt.xticks(rotation=45, ha='right',color='red')
print()
```

```python
import numpy as np
from numpy import *
data=np.c_[min_arbre.values,max_arbre.values]
name=np.r_[min_arbre.columns,max_arbre.columns]
data2=pd.DataFrame(data,columns=name)
```

# Creation d'une boucle sur la colonne 'HAUTEUR' pour creer une nouvelle     colonne 'hauteur_classe_col' en trois categories :
    ## 1: 0-15 M
    ## 2: 16-25 M
    ## 3: +25  M

```python
hauteur_classe = []

for i in df_new.HAUTEUR:
    if i >= 1 and i < 16:
        hauteur_classe.append('0-15 M')
    elif i >= 16 and i < 26:
        hauteur_classe.append('16-25 M')
    else:
        hauteur_classe.append('+25 M')

df_new['hauteur_classe_col']=[i for i in hauteur_classe]

df_new['hauteur_classe_col'].head()
```

# Je supprime toutes les colonnes inutiles avec la fonction DROP

```python
#je supprime toutes les colonnes inutiles
df_new.drop(['TYPEEMPLACEMENT','DOMANIALITE','COMPLEMENTADRESSE','LIEU / ADRESSE','IDEMPLACEMENT','NUMERO','VARIETEOUCULTIVAR','CIRCONFERENCEENCM','STADEDEVELOPPEMENT','IDBASE','REMARQUABLE'], axis = 1)
```

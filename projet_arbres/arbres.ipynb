```python
#Pou pouvoir créer des dataframes
import pandas

#Pour utiliser les données de l'index.csv
import csv

#Numpy sert à manipuler des matrices, tableaux multidimensionnels et fonctions mathématiques opérant sur ces tableaux
import numpy

#Donne un alias à numpy
import numpy as np

#Donne un alias à seaborn, utilisé pou les couleurs de diagrammes
import seaborn as sns
sns.set()

#Matplotlib est une bibliothèque du langage Python destinée à tracer et visualiser des données sous formes de graphiques
import matplotlib

#donne un alias à matplotlib.mlab
import matplotlib.mlab as mlab

# change le style des graphes
import matplotlib.pyplot as plt

# indique à Jupyter d'insérer les graphiques dans le notebook plutôt que dans une fenêtre externe
"%matplotlib inline"

#Création d'un Dataframe "df" avec les données du fichier.csv
df = pandas.read_csv('les-arbres.csv', sep = ';', header = 0)

#Nettoyage du dataframe en enlevant les arbres ayant une hauteur inférieur à 1 et supérieur à 40 mètres stocké dans "df_new".
df_new =  df[(df['HAUTEUR']>=1) & (df['HAUTEUR']<=40)]

#Visualisation du nombre de ligne et de colonnes du dataframe "df_new"
df_new.shape
```

```python
#Q1.1. Quels arrondissement qui compte le plus d’arbres ? 

#stockage de la première valeur du tableau "df_new" rangé dans l'ordre croissant, par arrondissements et compté par leur "IDBASE" dans "max_min_arrond".
max_min_arrond = df_new.groupby(["ARRONDISSEMENT"]).count().sort_values("IDBASE").tail(1)
#préparation affichage par arrondissement et hauteur max
max_arrond = max_min_arrond.groupby(['ARRONDISSEMENT','HAUTEUR']).max()
#Retait des colonnes superflues
resultat_max = max_arrond.drop(['IDBASE','TYPEEMPLACEMENT','DOMANIALITE','COMPLEMENTADRESSE','NUMERO','LIEU / ADRESSE','IDEMPLACEMENT','LIBELLEFRANCAIS', 'GENRE', 'ESPECE', 'VARIETEOUCULTIVAR', 'CIRCONFERENCEENCM','STADEDEVELOPPEMENT','REMARQUABLE','geo_point_2d'], axis=1)
#Affichage
resultat_max
```

```python
#Q1.2. Quel  arrondissement qui compte le moins d’arbres ?

#stockage de la première valeur du tableau "df_new" rangé dans l'ordre décroissant, par arrondissements et compté par leur "IDBASE" dans "max_min_arrond".
max_min_arrond = df_new.groupby(["ARRONDISSEMENT"]).count().sort_values("IDBASE", ascending = False).tail(1)
#préparation affichage par arrondissement et hauteur min
min_arrond = max_min_arrond.groupby(['ARRONDISSEMENT','HAUTEUR']).min()
#Retait des colonnes superflues
resultat_min = min_arrond.drop(['IDBASE','TYPEEMPLACEMENT','DOMANIALITE','COMPLEMENTADRESSE','NUMERO','LIEU / ADRESSE','IDEMPLACEMENT','LIBELLEFRANCAIS', 'GENRE', 'ESPECE', 'VARIETEOUCULTIVAR', 'CIRCONFERENCEENCM','STADEDEVELOPPEMENT','REMARQUABLE','geo_point_2d'], axis=1)
#Affichage
resultat_min

```

```python
arbres_arr = df_new.groupby(['ARRONDISSEMENT']).size()
plt.figure(figsize=(20,10))
arbres_arr = df_new.groupby(['ARRONDISSEMENT']).size().reset_index(name='Nombre d arbres')
plt.bar(arbres_arr['ARRONDISSEMENT'],arbres_arr['Nombre d arbres'],color='pink')
plt.xticks(rotation=45, ha='right',color='black')
plt.title('NOMBRE ARBRES PAR ARRONDISSEMENT')
print()
```

```python
arbres_arr = df_new.groupby(['ARRONDISSEMENT']).size().reset_index(name='Nombre d arbres')

plt.rcParams['figure.figsize']=(10,10)

df_new.ARRONDISSEMENT.value_counts().plot.pie()

plt.legend(prop={'size':10},loc = 'center')
plt.title('NOMBRE D\'ARBRES PAR ARRONDISSEMENT')
plt.ylabel("");
```

```python
# Q2.1. Dans quelle zone trouve-t-on les arbres les plus hauts ?

#stockage des valeurs les plus hautes du tableau "df_new" par "arrondissements" et "hauteur" dans "df_haut_max_min".
df_haut_max_min = df_new.groupby(['ARRONDISSEMENT','HAUTEUR']).max()
#stockage des valeurs du tableau "df_haut_max_min" rangé dans l'ordre décroissant, par "arrondissements" et "hauteur min".
df_haut_max = df_haut_max_min.sort_values("HAUTEUR", ascending = False)
#Affichages des 20 premiers
df_haut_max.head(20)
```

```python
plt.figure(figsize=(20,10))
max_arbre=df_new.groupby(['ARRONDISSEMENT']).HAUTEUR.max().reset_index(name='taille maximum')
plt.bar(max_arbre['ARRONDISSEMENT'],max_arbre['taille maximum'],color='salmon')
plt.xticks(rotation=45, ha='right',color='black')
print()
```

```python
# Q2.2. Dans quelle zone trouve-t-on les arbres les plus petits ?

#stockage des valeurs les plus basses du tableau "df_new" par "arrondissements" et "hauteur" dans "df_haut_max_min".
df_haut_min = df_new.groupby(['ARRONDISSEMENT','HAUTEUR']).min()
#stockage des valeurs du tableau "df_haut_max_min" rangé dans l'ordre croissant, par "arrondissements" et "hauteur min".
df_haut_min = df_haut_max_min.sort_values("HAUTEUR")
#Affichages des 20 premiers
df_haut_min.head(20)
```

```python
#hauteur minimum par arrondissement 
plt.figure(figsize=(20,10))
min_arbre = df_new.groupby(['ARRONDISSEMENT']).HAUTEUR.min().reset_index(name='taille minimum')
plt.bar(min_arbre['ARRONDISSEMENT'],min_arbre['taille minimum'],color='orange')
plt.xticks(rotation=45, ha='right',color='red')
print()
```

```python
hauteur_classe = []

for i in df_new.HAUTEUR:
    if i >= 1 and i < 16:
        hauteur_classe.append('0-15 M')
    elif i >= 16 and i < 26:
        hauteur_classe.append('16-25 M')
    else:
        hauteur_classe.append('+25 M')

df_new['hauteur_classe_col']=[i for i in hauteur_classe]

df_new['hauteur_classe_col'].head()
plt.rcParams['figure.figsize']=(20,20)
for i,t in enumerate (df_new.ARRONDISSEMENT.unique()):
    plt.subplot(7,4,(i+1))
    plt.title(t)
    data=df_new[df_new['ARRONDISSEMENT']==t]
    data.hauteur_classe_col.value_counts().plot.pie()
    plt.ylabel(" ")
```

```python
haut_arrond=df_new.groupby(['ARRONDISSEMENT']).hauteur_classe_col.value_counts().reset_index(name='total')
plt.rcParams['figure.figsize']=(10,10)
taille_3=haut_arrond[haut_arrond['hauteur_classe_col']=="+25 M"]
df_new.hauteur_classe_col.value_counts().plot.pie()
plt.legend(prop={'size':15})
plt.title('HAUTEUR DES ARBRES A ¨PARIS' )
plt.ylabel("");
```

```python
# Q3. Quels types d'arbres sont plantés dans les arrondissements de Paris ?

#stockage du nombre "d'espèces" par "arondissements" avec création d'une colonne "total" dans "espece_arr".
espece_arr = df_new.groupby(['ARRONDISSEMENT','ESPECE']).size().reset_index(name="TOTAL")
#stockage des valeurs obtenues dans un nouveau datafame "espece_par_arr" en retirant les lignes qui ne contiennent pas "PARIS" dans la colonne "ARRONDISSEMENTS".
espece_par_arr = espece_arr[espece_arr['ARRONDISSEMENT'].str.contains(pat = "PARIS")]
#affichage
espece_par_arr
```

```python
#Q3. Quels espèces d’arbres sont les plus répandues? 

espece_arr = df_new.groupby(["ESPECE"]).count().sort_values("IDBASE").tail(10)

espece_arr_max = espece_arr.groupby(['ESPECE','ARRONDISSEMENT']).max()

resultat_espece_max = espece_arr_max.drop(['IDBASE','TYPEEMPLACEMENT','DOMANIALITE','COMPLEMENTADRESSE','NUMERO','LIEU / ADRESSE','IDEMPLACEMENT','LIBELLEFRANCAIS', 'GENRE', 'VARIETEOUCULTIVAR', 'CIRCONFERENCEENCM', 'HAUTEUR', 'STADEDEVELOPPEMENT','REMARQUABLE','geo_point_2d'], axis=1)

resultat_espece_max
```

```python
#Q3. Quels espèces d’arbres sont les plus répandues? 

espece_arr = df_new.groupby(["ESPECE"]).count().sort_values("IDBASE", ascending = False).tail(10)

espece_arr_min = espece_arr.groupby(['ESPECE','ARRONDISSEMENT']).min()

resultat_espece_min = espece_arr_min.drop(['IDBASE','TYPEEMPLACEMENT','DOMANIALITE','COMPLEMENTADRESSE','NUMERO','LIEU / ADRESSE','IDEMPLACEMENT','LIBELLEFRANCAIS', 'GENRE', 'VARIETEOUCULTIVAR', 'CIRCONFERENCEENCM', 'HAUTEUR', 'STADEDEVELOPPEMENT','REMARQUABLE','geo_point_2d'], axis=1)

resultat_espece_min
```

```python
import pandas

import folium
df = pandas.read_csv('les-arbres.csv', sep = ';', header = 0)
df_new =  df[(df['HAUTEUR']>=1) & (df['HAUTEUR']<=40)]

max_min_arrond = df_new.groupby(["ARRONDISSEMENT"]).count().sort_values("IDBASE", ascending = False).tail(1)
min_arrond = max_min_arrond.groupby(['ARRONDISSEMENT','HAUTEUR']).min()
resultat_min = min_arrond.drop(['IDBASE','TYPEEMPLACEMENT','DOMANIALITE','COMPLEMENTADRESSE','NUMERO','LIEU / ADRESSE','IDEMPLACEMENT','LIBELLEFRANCAIS', 'GENRE', 'ESPECE', 'VARIETEOUCULTIVAR', 'CIRCONFERENCEENCM','STADEDEVELOPPEMENT','REMARQUABLE'], axis=1)
resultat_min

carte = folium.Map(location=[48.866667, 2.333333],zoom_start=10)

for i in max_min_arrond:
    folium.Marker([max_min_arrond['geo_point_2d']]).add_to(carte)

carte.save('index.html')

carte
```
